#1.**Build a simple webpage that displays text as shown in the below image**



```python
%%html
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>simple webpage</title>
    <body>
      <h1><b>This text will be bolded</b></h1>
      <p><i>This text will be italic </i></p>
      <p><u>This tag will be underlined</u></p>
      <p><mark>This tag will be highlighted</mark></p>
      <p>
        This is normal text <sup>This will be super scripted</sup>This is normal
        again
      </p>
      <p>This is normal text <sub>This tag will be subscripted</sub></p>
      <p><big>Normal Text</big> <small>Smal Text</small></p>
      <p><del>This tag will be deleted</del></p>
    </body>
  </head>
</html>
```

#**5.Create a description list of full stack web development tech stack, using the <dl> tag. Each term should be a tech stack name and each description should be a brief explanation of what the tech stack is used for**

```python
%%html
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>description list:explain</title>
  </head>
  <body>
    <h1>Full Stack Web Development Tech Stack</h1><hr>
    <h2>Frontend Technologies:</h2>
    <dl>
      <dt>
        HTML (HyperText Markup Language)
      </dt>
      <dd>HTML is the standard language for creating web pages. It provides the basic structure of a website using elements like headings, paragraphs, images, links, and forms.</dd>
      <dt>CSS (Cascading Style Sheets) </dt>
      <dd>CSS is used for designing and styling web pages. It controls the layout, colors, fonts, and responsiveness of a website to ensure an attractive and user-friendly interface.</dd>
      <dt>JavaScript </dt>
      <dd>JavaScript is a dynamic programming language that enables interactive elements on websites, such as animations, pop-ups, and real-time updates. It is essential for frontend development and is supported by all modern browsers.</dd>
      <dt>React.js</dt>
      <dd>React is a popular JavaScript library developed by Facebook for building interactive and reusable UI components. It follows a component-based architecture and uses a virtual DOM to optimize rendering performance.</dd>
      <dt>Angular</dt>
      <dd>Angular is a TypeScript-based web application framework developed by Google. It is used to build scalable and maintainable single-page applications (SPAs) with built-in dependency injection and two-way data binding.</dd>
      <h2>Backend Technologies:</h2>
      <dt>Node.js </dt>
      <dd>Node.js is a runtime environment that allows JavaScript to be executed on the server side. It is commonly used for building scalable and fast backend applications with event-driven architecture.</dd>
      <dt>Express.js</dt>
      <dd>Express.js is a lightweight and flexible web framework for Node.js. It simplifies backend development by providing tools for handling routes, middleware, and server requests.

      </dd>
        <dt>Django</dt>
        <dd>Django is a high-level Python web framework that promotes rapid development and clean, pragmatic design. It follows the MVC (Model-View-Controller) architecture and includes built-in security features.</dd>
        <h2>Databases</h2>
        <dt>MongoDB</dt>
        <dd>MongoDB is a NoSQL database that stores data in JSON-like documents. It is highly scalable and suitable for applications requiring high performance and flexibility.</dd>
        <dt>MySQL</dt>
        <dd>MySQL is a widely used open-source relational database that follows the SQL (Structured Query Language) standard. It is known for its reliability, efficiency, and ease of use.</dd>
        <dt>PostgreSQL</dt>
        <dd>PostgreSQL is an advanced open-source relational database that supports complex queries, data integrity, and high performance. It is widely used in enterprise applications.</dd>
        <h2>Version Control & Deployment</h2>
        <dt>Git & GitHub</dt>
        <dd>Git is a distributed version control system used to track code changes, while GitHub is a cloud-based platform for code collaboration, hosting, and version management.

        </dd>
        <dt>Docker</dt>
        <dd>Docker is a containerization platform that allows developers to package applications along with their dependencies to ensure consistency across different environments.</dd>
        <dt>AWS (Amazon Web Services)</dt>
        <dd>AWS is a cloud computing platform that provides various services for hosting, computing, storage, and networking, allowing developers to deploy and scale web applications easily.</dd>

    </dl>
  </body>

</html>
```
